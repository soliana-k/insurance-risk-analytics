#### 2. What are the distributions of key financial variables? Are there outliers in TotalClaims or CustomValueEstimate that could skew analysis?

##### Distributions of Key Financial Variables

All four financial variables (`TotalPremium`, `TotalClaims`, `CapitalOutstanding`, and `CustomValueEstimate`) exhibit **extreme positive skewness**. The visualizations show an overwhelming concentration of data points clamped right at or near zero, trailing off into incredibly long, sparse right-hand tails.

##### CustomValueEstimate

* **The Problem:** The IQR calculation shows an **IQR of 0.00** and an **Upper Whisker Limit of 0.00**. This happens mathematically because more than 75% of the values in this column are exactly zero (or imputed zeros).
* **The Impact:** Because the statistical baseline threshold defaults to zero, any policy containing an actual custom vehicle valuation is flagged as an anomaly. This labels a staggering **22.06% of the dataset (220,456 records)** as outliers.
* **Handling Strategy:** Treating nearly a quarter of the data as standard anomalies is a mistake. Instead, handle this feature downstream using **feature engineering** (e.g., using the binary `Has_Custom_Estimate` indicator or evaluating values conditionally where `CustomValueEstimate > 0`).

##### TotalPremium

* **The Problem:** With a tight IQR of 21.93, any policy premium exceeding **54.82** is flagged as an anomaly, capturing **20.89% of the data (208,836 records)** as outliers.
* **The Impact:** The chart shows a tiny fraction of massive fleet policies or heavy commercial coverages extending out to over 2,000 currency units. These few massive premiums will skew simple linear regressions heavily.

##### TotalClaims

* **The Problem:** Similar to custom estimates, the vast majority of records are zero claims, forcing an **IQR and Upper Whisker of 0.00**. This flags **2,775 rows (0.278%)** as outliers.
* **The Impact:** Even though the percentage is small, look at the x-axis on the distribution plot: the tail stretches all the way out to **400,000+ units**. These represent catastrophic, high-severity insurance losses.



#### 3. Are there temporal trends? Did claim frequency or severity change over the 18-month period?

An analysis of the 18-month historical timeline reveals a distinct structural relationship between portfolio performance metrics, characterized by a stable operating plateau followed by a severe data truncation anomaly at the terminal edge of the observation window. Throughout the core tracking period from early 2014 through mid-2015, monthly claim frequency maintained a steady, elevated run-rate fluctuating between **0.30% and 0.37%**, while average claim severity remained systematically contained within a predictable corridor of **21,000 to 26,000** currency units. However, this historical equilibrium diverges violently in August 2015, where the claim frequency rate collapses to a near-zero absolute low of **0.02%** while average claim severity spikes to an uncharacteristic historical peak of **46,355.41** units. Rather than signaling a sudden shift in underlying underwriting risk, this extreme late-stage decoupling indicates a classic insurance reporting lag or Incurred But Not Reported (IBNR) data maturity constraint, wherein standard low-severity claims remain open or unlogged at the close of the data extraction window, leaving only isolated, high-severity catastrophic losses fully committed to the database.
