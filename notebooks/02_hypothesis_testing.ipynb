
### Executive Hypothesis Testing Report

#### 1. Summary of Decisions

* **Hypothesis 1 (Province vs. Claim Frequency):** **REJECT $H_0$**. Location significantly impacts how often claims are made.
* **Hypothesis 2 (Zip Code vs. Claim Severity):** **FAIL TO REJECT $H_0$**. Zip codes do not change the average cost of a claim.
* **Hypothesis 3 (Zip Code vs. Net Margin):** **FAIL TO REJECT $H_0$**. There is no significant profit difference between the compared zip codes.
* **Hypothesis 4 (Gender vs. Claim Frequency):** **REJECT $H_0$**. Men claim significantly more often than women.



#### 2. Findings & Business Interpretations

##### Hypothesis 1: Regional Risk (Claim Frequency)

* **What the data shows:** The claim frequency in the Western Cape is **0.2093%**, while Gauteng is significantly higher at **0.3359%** ($p = 1.38 \times 10^{-15}$).
* **Observation:** Drivers in Gauteng are roughly **60% more likely** to file a claim than drivers in the Western Cape. This is a massive, real risk difference.
* **Business Move:** Stop charging the same base rate across the country. Charge higher base premiums for vehicles operating in Gauteng to cover the higher accident rates.

##### Hypothesis 2: Geographic Risk (Claim Severity)

* **What the data shows:** The average claim cost in Postal Code 2000 is **19,196.41**, compared to **18,162.03** in Postal Code 122 ($p = 0.700$).
* **Observation:** The ~1,000 unit difference is completely due to random statistical noise. Once an accident happens, the cost to repair or replace the vehicle is structurally the same in both areas.
* **Business Move:** Do not split your claims or repair budgets by specific zip codes. Treat urban accident costs as uniform.

##### Hypothesis 3: Geographic Performance (Net Margin)

* **What the data shows:** The average net margin per policy is **-8.16** for Postal Code 2000 and **-22.86** for Postal Code 122 ($p = 0.246$).
* **Observation:** Both of your highest-volume zip codes are actively losing money. The statistical test proves that neither zip code is performing significantly better than the other—they are both trapped in an underperforming deficit.
* **Business Move:** This is not a localized problem with one "bad" zip code. It is a systemic pricing issue. You need a blanket premium increase across high-volume urban policies to push the margins back into the positive.

##### Hypothesis 4: Demographic Risk (Gender Claim Frequency)

* **What the data shows:** By using customer titles to bypass the missing data, we analyzed 99.92% of the portfolio. Reconstructed women have a claim frequency of **0.1978%**, while reconstructed men sit at **0.2833%** ($p = 6.65 \times 10^{-5}$).
* **Observation:** Men are **43% more likely** to file a claim than women. Because this was tested across almost the entire dataset, the gap is definitive.
* **Business Move:** Gender is a clear risk driver for how often a claim occurs. Incorporate this frequency gap into your segmentation strategy to price male drivers accurately according to their higher likelihood of filing claims.
